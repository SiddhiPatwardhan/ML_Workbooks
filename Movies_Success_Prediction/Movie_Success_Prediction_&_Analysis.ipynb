```python
def convert(obj):
    L =[]
    for i in ast.literal_eval(obj):
        i['name']
        L.append(i['name'])
    return L
```

```python
import ast
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['production_companies'] = movies['production_companies'].apply(convert)
movies.head()
```

```python
movies['production_countries'] = movies['production_countries'].apply(convert)
movies.head()
```

```python
movies['spoken_languages'] = movies['spoken_languages'].apply(convert)
movies.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
# remove the spaces -> bcoz it may lead to confusions in model
# Apply this transformation
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['production_companies'] = movies['production_companies'].apply(lambda x:[i.replace(" ","")for i in x])
movies['production_countries'] = movies['production_countries'].apply(lambda x:[i.replace(" ","")for i in x])
movies['spoken_languages'] = movies['spoken_languages'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
movies.to_csv('movies.csv',index=False)
```

```python
# First, clean 0 or missing values
movies = movies[(movies['budget'] > 1000000) & (movies['revenue'] > 0)]
```

```python
movies.shape
```

```python
# Label: Hit (1) if profit margin â‰¥ 20%
movies['success'] = (movies['revenue'] > movies['budget'] * 1.2).astype(int)
```

```python
# Convert release_date to datetime
movies['release_date'] = pd.to_datetime(movies['release_date'], errors='coerce')
movies['year'] = movies['release_date'].dt.year
```

```python
movies.head()
```

```python
from collections import Counter
# ---- 1. Top 10 Genres ----
genre_count = Counter([genre for genres in movies['genres'] for genre in genres])
top_genres = dict(genre_count.most_common(10))

plt.figure(figsize=(10, 6))
plt.bar(top_genres.keys(), top_genres.values(), color='skyblue')
plt.title("Top 10 Genres")
plt.xticks(rotation=45)
plt.ylabel("Number of Movies")
plt.show()
```

```python
# ---- 2. Revenue vs Budget ----
plt.figure(figsize=(8, 6))
sns.scatterplot(data=movies, x='budget', y='revenue', hue='success', palette='coolwarm')
plt.title("Budget vs Revenue (Color: Success)")
plt.show()
```

```python
# ---- 3. Hit vs Flop Count ----
plt.figure(figsize=(6, 5))
sns.countplot(x='success', data=movies)
plt.xticks([0, 1], ['Flop', 'Hit'])
plt.title("Hit vs Flop Distribution")
plt.show()
```

```python
# ---- 4. Average Revenue by Genre ----
df_exploded = movies.explode('genres')
df_exploded['genres'] = df_exploded['genres'].apply(lambda x: x.strip("[]").replace("'", "").replace('"', '') if isinstance(x, str) else x)
avg_revenue = df_exploded.groupby('genres')['revenue'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
avg_revenue.plot(kind='bar', color='orange')
plt.title("Top 10 Genres by Avg Revenue")
plt.ylabel("Average Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
# ---- 5. Top 10 Actors ----
actor_count = Counter([actor for cast in movies['cast'] for actor in cast])
top_actors = dict(actor_count.most_common(10))

plt.figure(figsize=(10, 6))
plt.bar(top_actors.keys(), top_actors.values(), color='purple')
plt.title("Top 10 Most Frequent Actors")
plt.xticks(rotation=45)
plt.ylabel("Number of Movies")
plt.show()
```

```python
# ---- 6. Hits Over the Years ----
hits_per_year = movies[movies['success'] == 1].groupby('year').size()

plt.figure(figsize=(10, 6))
hits_per_year.plot(marker='o')
plt.title("Number of Hit Movies Per Year")
plt.xlabel("Year")
plt.ylabel("Hits")
plt.grid(True)
plt.show()
```
